```python
from datetime import date,timedelta
import pandas as pd
import numpy as np
import pickle
with open("all_of_dic.pkl", "rb") as file:
    all_df_dic = pickle.load(file)
```

# 特定投資株式，みなし保有株式を取り出す

```python
# 特定投資株式
col_specified = ["NameOfSecuritiesDetailsOfSpecifiedInvestmentEquitySecuritiesHeldForPurposesOtherThanPureInvestmentReportingCompany",
              "PurposeOfShareholdingQuantitativeEffectsOfShareholdingAndReasonForIncreaseInNumberOfSharesDetailsOfSpecifiedInvestmentSharesHeldForPurposesOtherThanPureInvestmentReportingCompany",
              "WhetherIssuerOfAforementionedSharesHoldsReportingCompanysSharesDetailsOfSpecifiedInvestmentSharesHeldForPurposesOtherThanPureInvestmentReportingCompany",
              "NumberOfSharesHeldDetailsOfSpecifiedInvestmentEquitySecuritiesHeldForPurposesOtherThanPureInvestmentReportingCompany",
              "BookValueDetailsOfSpecifiedInvestmentEquitySecuritiesHeldForPurposesOtherThanPureInvestmentReportingCompany"]
# みなし保有株式
col_deemed = ["NameOfSecuritiesDetailsOfDeemedHoldingsOfEquitySecuritiesHeldForPurposesOtherThanPureInvestmentReportingCompany",
       "PurposeOfShareholdingQuantitativeEffectsOfShareholdingAndReasonForIncreaseInNumberOfSharesDetailsOfDeemedHoldingsOfSharesHeldForPurposesOtherThanPureInvestmentReportingCompany",
       "WhetherIssuerOfAforementionedSharesHoldsReportingCompanysSharesDetailsOfDeemedHoldingsOfSharesHeldForPurposesOtherThanPureInvestmentReportingCompany",
       "NumberOfSharesHeldDetailsOfDeemedHoldingsOfEquitySecuritiesHeldForPurposesOtherThanPureInvestmentReportingCompany",
       "BookValueDetailsOfDeemedHoldingsOfEquitySecuritiesHeldForPurposesOtherThanPureInvestmentReportingCompany"]

col_list = ["name_of_securities", "purpose", "held", "number", "book_value"]
```

```python
df_capital_tie = pd.DataFrame(columns = ["code", "type"] + col_list)

for idx, code in enumerate(all_df_dic["有価証券報告書"].keys()):
    print(idx, "/", len(code_list), end = "\r")
    print(code, end = "\r")
    df_idx = all_df_dic["有価証券報告書"][code]
    df_idx = df_idx[df_idx["提出日"] == max(df_idx["提出日"].unique())]
    # print("specified")
    # 特定投資株式
    df_idx_specified = df_idx[df_idx["tag"].isin(col_specified)].pivot(index = "context", columns = "tag", values = "値").dropna(how = "any")
    df_idx_specified.rename(columns = dict(zip(col_specified, col_list)), inplace = True)
    if len(df_idx_specified) != 0:
        df_capital_tie = df_capital_tie.append(df_idx_specified.sort_values(by = "book_value", ascending = False))
        df_capital_tie.fillna({"code": code, "type": "specified"}, inplace = True)
    
    # print("deemed")
    # みなし保有株式
    df_idx_deemed = df_idx[df_idx["tag"].isin(col_deemed)].pivot(index = "context", columns = "tag", values = "値").dropna(how = "any")
    df_idx_deemed.rename(columns = dict(zip(col_deemed, col_list)), inplace = True)
    if len(df_idx_deemed) != 0:
        df_capital_tie = df_capital_tie.append(df_idx_deemed.sort_values(by = "book_value", ascending = False))
        df_capital_tie.fillna({"code": code, "type": "deemed"}, inplace = True)

df_capital_tie.to_excel("capital_20220130.xlsx", index = False)
```
